# 第四题：数位递增的数
问题：一个正整数如果任何一个数位不大于右边相邻的数位，则称为一个数位递增的数，例如1135是一个数位递增的数，而1024不是一个数位递增的数。
给定正整数 n，请问在整数 1 至 n 中有多少个数位递增的数？


```python
def check(n):
    n = list(str(n))
    m = n.copy()
    m.sort()
    if n == m:
        return True
res = check(1024)
print(res)
```

```python
def check_1(n):
    n = str(n)
    for i in range(len(n)-1):
        if n[i]>n[i+1]:
            return False
    return True
ans = 0
n = int(input('请输入一个正整数：'))
for i in range(1,n+1):
    ret = check_1(i)
    if ret:
        ans += 1
print(ans)
```

```python
def check_2(n):
    n = list(str(n))
    m = n.copy()
    m.sort()
    if n == m:
        return True
ans = 0
n = int(input('请输入一个正整数：'))
for i in range(1,n+1):
    ret = check_2(i)
    if ret:
        ans += 1
print(ans)
```
