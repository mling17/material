### 2.5 字符串拼接

str1+str2
'我的名字是%s，年龄是%s'%('tom',18)
'我的名字是{}，年龄是{}'.format('tom',18)
'我的名字是{0}，年龄是{1}'.format('tom',18)
'我的名字是{name}，年龄是{age}'.format(name='tom',age=18)
f'我的名字是{name},年龄是{age}'


```python
name = 'Tom'
age = 18
#'我的名字是Tome，年龄18岁'
```

```python
# 1. +
print('我的名字是' + name+'，年龄'+str(age)+'岁')
```

```python
# 2. %s
print('我的名字是%s，年龄%d岁'%(name,age)) # 推荐
```

```python
# 3. format方法
print('我的名字是{}，年龄{}岁'.format(name,age))
```

```python
# 4. format方法
print('我的名字是{1}，年龄{0}岁,她和我重名也叫{1}'.format(age,name))
```

```python
# 5. format方法
'我的名字是{name}，年龄是{age}'.format(age=age,name=name)
```

```python
# 6. f,python3.6以上版本支持
f'我的名字是{name},年龄是{age}'#推荐
```
