# os库
os库主要对目标和文件操作

## 常用模块方法

```python
import os
os.getcwd() # pwd
os.chdir('/tmp') # cd /tmp
os.listdir('/tmp') # ls /tmp
os.makedirs('/tmp/mydemo/mydir') # mkdir -p 创建多级目录
os.mkdir('/tmp/dir') # mkdir 创建单个目录
os.rmdir('/tmp/dir') # rm dir 删除目录
os.mknod('hello') # touch hello
os.symlink('/etc/hosts','zhuji') # ln -s /etc/hosts zhuji
os.chmod('hello',0o644) # chmod 644 hello
os.rename('hello','welcome') # mv hello welcome
os.unlink('zhuiji') # 删除软连接
os.remove('welcome') # rm -f welcome 删除文件
```

## os.path子模块

```python
import os
os.path.abspath('.') # 当前文件夹/文件的绝对路径
```

```python
os.path.dirname('D:\material\README.md') # 获取文件/文件夹所在的目录
```

```python
os.path.basename('D:\material\README.md') # 获取文件/文件夹的名字
```

```python
os.path.join('D:\material','README.md') # 拼接路径
```

```python
print(os.path.split('D:\material\README.md')) # 分割路径为一个元组
print(os.path.split('D:\material'))
```

```python
os.path.splitext('D:\material\README.md') # 分割文件后缀为一个元组
```

```python
print(os.path.isdir('D:\material\README.md')) # 判断路径是否是一个目录
print(os.path.isdir('D:\material'))
```

```python
print(os.path.isfile('D:\material\README.md')) # 判断路径是否是一个文件
print(os.path.isfile('D:\material'))
```

```python
print(os.path.islink('D:\material\README.md')) # 判断路径是否是一个链接
```

```python
print(os.path.exists('D:\material\README.md')) # 判断路径是否是存在
```

# subprocess
subprocess库用于执行Shell命令，工作时会fork一个子进程去执行任务，连接到子进程的标准输入、输出、错误，并获得它们的返回代码。

这个模块将取代os.system、os.spawn*、os.popen*、popen2.和commands.。

subprocess的主要方法：
subprocess.run()，subprocess.Popen(),subprocess.call

语法：subprocess.run(args, \*, stdin=None, stdout=None, stderr=None, shell=False, cwd=None, timeout=None, check=False, encoding=None)

```python
import subprocess
rc = subprocess.run('png 127.0.0.1 -n 1',shell=True) # 这是个错误命令
print(rc.returncode)# 返回值为1
rc1 = subprocess.run('ping 127.0.0.1 -n 1',shell=True) # 这是个正确命令
print(rc1.returncode)# 命令执行成功返回值为0
```

# pickle

pickle模块实现了对一个Python对象结构的二进制序列化和反序列化。

主要用于将对象持久化到文件存储。

pickle模块主要有两个函数：
• dump() 把对象保存到文件中（序列化），使用load()函数从文件中读取（反序列化）
• dumps() 把对象保存到内存中，使用loads()函数读取

# json

JSON是一种轻量级数据交换格式，一般API返回的数据大多是JSON、XML，如果返回JSON的话，需将获取的数据转换成字典，方面在程序中处理。

json与pickle有相似的接口，主要提供两种方法：
• dumps() 对数据进行编码
• loads() 对数据进行解码
