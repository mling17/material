持久化：
- 基于终端指令
将parse方法的返回值保存道文本文件中
简介高效便捷，局限性强，只能保存到指定后缀的文件中
- 基于管道
    - 数据解析
    - 在item类中定义相关属性
    - 将解析的数据封装存储道item类型的对象
    - 将item类型的对象提交给管道进行持久化存储的操作
    - 在管道类中的process_item中要将其接受道的item对象存储
    - 配置文件中开启管道
    
    好处：通用性强
